```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def theta1(sample):
    return np.abs(theta - 2*np.mean(sample))

def theta2(sample):
    return np.abs(theta -np.min(sample))

def theta3(sample):
    return np.abs(theta -np.max(sample))

def theta4(sample):
    return np.abs(theta -np.max(sample) + np.min(sample))

def theta5(my_sample):
    sample = my_sample.copy()
    sample = np.sort(sample)
    if len(sample) == 1:
        return np.abs(theta - sample[0])
    else:
        return np.abs(theta - sample[0] - ((np.sum(sample[1:]))/(len(sample)-1)))
```

```python
x = []
theta = 200
y1, y2, y3, y4, y5 = [], [], [], [] ,[]
```

```python
my_sample = random.sample(range(1, theta), 100)
```

```python
for i in range(1,100):
    x.append(i)
    y1.append(theta1(my_sample[:i]))
    y2.append(theta2(my_sample[:i]))
    y3.append(theta3(my_sample[:i]))
    y4.append(theta4(my_sample[:i]))
    y5.append(theta5(my_sample[:i]))
```

```python
def plotter(x, y, label):
    fig, ax = plt.subplots(figsize=(9, 6))
    
    ax.set_xticks(np.linspace(0, np.max(x), 15).astype('int'))
    ax.set_yticks(np.linspace(0, np.max(y), 15).astype('int'))
    ax.grid(which="major",alpha=0.6)

    ax.plot(x, y, color = 'k', label = label)
    ax.legend(loc='best', prop={'size': 15})
```

```python
plotter(x,y1,'θ = 2*x̄')
```

```python
plotter(x, y2, 'θ = Xmin')
```

```python
plotter(x, y3, 'θ = Xmax')
```

```python
plotter(x, y4, 'θ = Xmax + Xmin')
```

```python
plotter(x, y5,  'θ = X1 + sum(X[1:])/(n-1)')
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax.set_title('which estimation converges?')

ax.set_xticks(np.linspace(0, 100, 15).astype('int'))
ax.set_yticks(np.linspace(0, theta, 15).astype('int'))
ax.grid(which="major",alpha=0.6)

ax.plot(x, y1, label = 'θ = 2*x̄')
ax.plot(x, y2, label = 'θ = Xmin')
ax.plot(x, y3, label = 'θ = Xmax')
ax.plot(x, y4, label = 'θ = Xmax + Xmin')
ax.plot(x, y5, label = 'θ = X1 + sum(X[1:])/(n-1)')

ax.legend(loc = 'center right', fancybox=True,  bbox_to_anchor=(1, 0.7))
```
